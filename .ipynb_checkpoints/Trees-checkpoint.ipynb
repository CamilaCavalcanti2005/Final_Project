```python
import pandas as pd

trees = pd.read_csv("citytrees.csv", low_memory=False)

trees.count()
```

After importing the pandas and bringing in the data CVS, we create a dictionary for each neighboorhood, first by inserting each neighboorhood into the dictionary with a value of 0, and loop through to assign different point values according to the the qualities of the many, many trees within the dataset.

```python
treePoints = dict()


for i, r in trees.iterrows():
    treePoints[r['neighborhood']] = 0

for i, r in trees.iterrows():
    if (r['condition']=="Dead"):
        treePoints[r['neighborhood']] = treePoints[r['neighborhood']] - 2 #0.0125  * 160
    elif(r['condition']=="Critical"):
        treePoints[r['neighborhood']] = treePoints[r['neighborhood']] - 1
    elif(r['condition']=="Fair"):
        treePoints[r['neighborhood']] = treePoints[r['neighborhood']] + 1
    elif(r['condition']=="Good"):
        treePoints[r['neighborhood']] = treePoints[r['neighborhood']] + 2
    elif(r['condition']=="Very Good"):
        treePoints[r['neighborhood']] = treePoints[r['neighborhood']] + 3
    elif(r['condition']=="Excellent"):
        treePoints[r['neighborhood']] = treePoints[r['neighborhood']] + 4
        


for key,value in treePoints.items():
    treePoints[key] = treePoints[key]/160
 
        
print(treePoints)


```

We have to make sure there is no values within the arrays of names and scores that are incompatible with the rest, like a na within the names, or else it will create an error when inputted into the graphing function.

```python
import matplotlib.pyplot as plt
%matplotlib inline


oldnames = (list(treePoints.keys()))
oldscores = list(treePoints.values())

names = []
scores = []

num =0
for x in oldnames:
    if type(x) == str:
        names.append(x)
        scores.append(oldscores[num])
    num = num + 1

num =0
for x in scores:
    scores[num] = scores[num]
    num = num + 1

plt.figure(figsize= (30, 6))
plt.bar(names, scores, color= 'forestgreen')
plt.xticks(rotation=45, ha='right')
plt.title('Points for Trees in each Neighboorhood')
plt.xlabel('Neighborhood')
plt.ylabel('Points')
plt.subplots_adjust(bottom=0.3)
plt.tight_layout()

plt.show()
```
