```python
import pandas as pd

trees = pd.read_csv("citytrees.csv")

trees.count()
```

```python
quality = dict()


for i, r in trees.iterrows():
    quality[r['neighborhood']] = 0

for i, r in trees.iterrows():
    if (r['condition']=="Dead"):
        quality[r['neighborhood']] = quality[r['neighborhood']] - 2 #0.0125  * 160
    elif(r['condition']=="Critical"):
        quality[r['neighborhood']] = quality[r['neighborhood']] - 1
    elif(r['condition']=="Fair"):
        quality[r['neighborhood']] = quality[r['neighborhood']] + 1
    elif(r['condition']=="Good"):
        quality[r['neighborhood']] = quality[r['neighborhood']] + 2
    elif(r['condition']=="Very Good"):
        quality[r['neighborhood']] = quality[r['neighborhood']] + 3
    elif(r['condition']=="Excellent"):
        quality[r['neighborhood']] = quality[r['neighborhood']] + 4
        
print(quality)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline


oldnames = (list(quality.keys()))
oldscores = list(quality.values())

names = []
scores = []

num =0
for x in oldnames:
    if type(x) == str:
        names.append(x)
        scores.append(oldscores[num])
    num = num + 1

num =0
for x in scores:
    scores[num] = scores[num]/160
    num = num + 1

print(names)
print(scores)

plt.figure(figsize= (30, 6))
plt.bar(names, scores, color= 'forestgreen')
plt.xticks(rotation=45, ha='right')
plt.show()

print(scores[2])
```
