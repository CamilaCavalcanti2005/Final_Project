Here I am counting the number of playgrounds in each neighborhood that has a playground. (Neighborhoods that had no playgrounds were not in the data set, and they won't be in consideration for points here anyway).

```python
# load pandas
import pandas as pd
import numpy as np

# Read the dataset into a DataFrame
df = pd.read_csv('playgrounds.csv')  # Replace 'your_dataset.csv' with the actual path to your dataset

# Count the occurrences of each neighborhood
neighborhood_counts = df['neighborhood'].value_counts()

# Print the result
print(neighborhood_counts)
```

I will now make a graph of the data:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
playgrounds_data = pd.read_csv("playgrounds.csv")

# Count the occurrences of each neighborhood
neighborhood_counts = playgrounds_data['neighborhood'].value_counts()

# Plotting
plt.figure(figsize=(12, 6))
neighborhood_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Playgrounds per Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Playgrounds')
plt.xticks(rotation=45, ha='right')
plt.subplots_adjust(bottom=0.3)  # Adjust spacing between x-ticks
plt.tight_layout()
plt.show()
```

As you can see, most of the neighborhoods only had 1 or 2 playgrounds. Squirrel Hill South has the most playgrounds by far, with 8 playgrounds. Beechview and South Side Slopes are the runners-up with 5 playgrounds each.
